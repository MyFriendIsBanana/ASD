### 1. Надати визначення складності задачі із символом Ω.

Символ Ω (омега) описує **нижню межу** складності алгоритму. Це означає, що алгоритм завжди виконується не повільніше, ніж визначено функцією Ω.

### 2. Функція часової складності має вигляд: F(N) = N^3 + 7N^2 - 14N. Як записати асимптотичну складність у O-нотації?

Для функції F(N) = N^3 + 7N^2 - 14N, ми залишаємо лише найбільший член з найвищим степенем, оскільки він визначає поведінку функції при великих значеннях N.

**Відповідь:** O(N^3)

### 3. Функція часової складності має вигляд: F(N) = 1.01N + N^{10}. Як записати асимптотичну складність у O-нотації?

У цьому випадку найвищий степінь буде N^{10}.

**Відповідь:** O(N^{10})

### 4. Функція часової складності має вигляд: F(N) = N^{1.3} + 10 \log_2 N. Як записати асимптотичну складність у O-нотації?

Найшвидше зростає  N^{1.3} , тому складність алгоритму буде O(N^{1.3}).

**Відповідь:**  O(N^{1.3}) 

### 5. У чому полягає ідея розпаралелювання обчислень і для чого вона використовується?

Розпаралелювання обчислень дозволяє розподілити задачі між кількома процесорами для пришвидшення виконання.

### 6. Які існують способи підвищення обчислювальної швидкості алгоритмів?

Способи:

- Використання паралельних обчислень.

## Приклад для розрахунку часу виконання алгоритму сортування (наприклад, бульбашкова сортування)

```python
import time
import random

arr = random.sample(range(1, 1001), 1000)

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

start_time = time.time()
bubble_sort(arr)
end_time = time.time()

execution_time = end_time - start_time
execution_time
```

### Оцінка часу виконання

У цій лабораторній роботі ми використовували алгоритм бульбашкового сортування, який має часову складність \( O(N^2) \). Ми виміряли час виконання для випадкового масиву з 1000 елементів. Це дозволяє оцінити, як швидко алгоритм виконується при різних розмірах вхідних даних.


## Порівняння швидкості з іншими алгоритмами, наприклад, швидкого сортування

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

start_time = time.time()
quick_sort(arr)
end_time = time.time()

execution_time_quick = end_time - start_time
execution_time_quick
```

```python
import time
import random
import matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def time_sort_algorithm(algorithm, arr):
    start_time = time.time()
    algorithm(arr)
    end_time = time.time()
    return end_time - start_time

sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
bubble_times = []
quick_times = []

for size in sizes:
    arr_bubble = random.sample(range(1, 10001), size) 
    arr_quick = arr_bubble.copy() 
    bubble_times.append(time_sort_algorithm(bubble_sort, arr_bubble))
    quick_times.append(time_sort_algorithm(quick_sort, arr_quick))

plt.figure(figsize=(10, 6))
plt.plot(sizes, bubble_times, label="Bubble Sort", color="blue", marker="o")
plt.plot(sizes, quick_times, label="Quick Sort", color="red", marker="x")
plt.xlabel('Розмір масиву (N)')
plt.ylabel('Час виконання (секунди)')
plt.title('Порівняння часу виконання алгоритмів сортування')
plt.legend()
plt.grid(True)
plt.show()
```

# Підсумки лабораторної роботи

### 1. Визначення складності задач

- **Ω-нотація** описує **нижню межу** складності алгоритму. Вона показує, що алгоритм не може працювати швидше за певний час навіть при найкращих умовах.
  
- **О-нотація** використовується для визначення асимптотичної складності задачі, тобто як змінюється час виконання алгоритму при збільшенні розміру вхідних даних.

### 2. Оцінка часу виконання

- Було проведено порівняння двох алгоритмів сортування:
  - **Bubble Sort (Бульбашкова сортування)** має складність \( O(N^2) \), і його час виконання збільшується квадратично з ростом розміру масиву.
  - **Quick Sort (Швидка сортування)** має складність \( O(N \log N) \), і час виконання збільшується значно повільніше, майже лінійно.

### 3. Розпаралелювання обчислень

Розпаралелювання дозволяє значно пришвидшити виконання задач, розподіляючи обчислення на кілька процесорів. Алгоритми, які підтримують таку можливість, дозволяють скоротити час виконання при великих даних.

### 4. Способи підвищення обчислювальної швидкості

Основні методи:
- Використання ефективних алгоритмів.
- Паралельні обчислення.
- Використання більш швидких структур даних.

### 5. Висновки

В результаті лабораторної роботи було продемонстровано на практиці, як складність алгоритмів впливає на їх ефективність при різних розмірах вхідних даних. За допомогою побудованого графіка стало очевидним, що **бульбашкова сортування** працює значно повільніше за **швидку сортування** при збільшенні розміру масиву. Тому для великих наборів даних рекомендується використовувати швидші алгоритми, як-от **швидка сортування**.
