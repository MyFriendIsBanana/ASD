## Контрольні питання

### 1. Сформулювати задачу пошуку

Задача пошуку полягає в знаходженні елемента або кількох елементів у колекції даних. Це може бути пошук конкретного значення у масиві, списку, дереві або іншій структурі даних. Алгоритми пошуку оптимізують цей процес для зменшення часу, необхідного для знаходження елемента, залежно від розміру даних.

### 2. Які критерії можна використовувати для оцінювання ефективності алгоритмів пошуку?

Основними критеріями для оцінки ефективності алгоритмів пошуку є:
- **Часова складність**: Оцінка швидкості алгоритму (O-нотація).
- **Просторова складність**: Кількість пам'яті, яку використовує алгоритм.
- **Швидкість виконання**: Час, необхідний для знаходження елемента в залежності від розміру входу.
- **Стійкість до змін у даних**: Як швидко алгоритм може працювати на змінних або частково впорядкованих даних.

### 3. Лінійний пошук

Лінійний пошук є одним з найпростіших алгоритмів пошуку. Він полягає в тому, щоб перевірити кожен елемент колекції по черзі до знаходження шуканого елемента або завершення перевірки всіх елементів.

**Асимптотична складність лінійного пошуку**: 
- Найгірший випадок: O(N), де N — кількість елементів у масиві.

### 4. Бінарний (двійковий) пошук

Бінарний пошук працює тільки на відсортованих масивах. Він розділяє масив на дві частини і порівнює шуканий елемент з елементом в середині масиву. Якщо шукане значення менше середнього елемента, пошук продовжується в лівій половині, інакше — в правій половині.

**Принцип роботи**:
1. Перевірити середній елемент.
2. Якщо він є шуканим, завершити пошук.
3. Якщо шуканий елемент менший за середній, повторити пошук у лівій половині.
4. Якщо більший — у правій половині.

**Асимптотична складність бінарного пошуку**:
- Найгірший випадок: O(log N), де N — кількість елементів у масиві.

### 5. Переваги і недоліки бінарного пошуку порівняно з лінійним

**Переваги бінарного пошуку**:
- **Швидкість**: Бінарний пошук значно швидший за лінійний, оскільки має складність \(O(\log N)\), що значно краще за \(O(N)\) лінійного пошуку.
- **Ефективність на великих даних**: При великих масивах бінарний пошук є набагато ефективнішим.

**Недоліки**:
- Потрібно, щоб масив був відсортований.
- При кожному порівнянні потрібно визначити середній елемент.

### 6. Інші алгоритми пошуку

Окрім лінійного та бінарного пошуку існують й інші алгоритми:
- **Блочний пошук** — використовується для великих масивів, розбитих на блоки.
- **Індексований пошук** — створення додаткової структури даних для прискорення пошуку.
- **Пошук у дереві (наприклад, пошук у двійкових деревах)** — використовує структуру дерева для швидкого пошуку.

## Реалізація алгоритмів пошуку

### Лінійний пошук

```python
def linear_search(arr, target):
 for index, value in enumerate(arr):
     if value == target:
         return index
 return -1


arr = [2, 3, 5, 7, 11, 13, 17]
target = 7
result = linear_search(arr, target)
print(f"Лінійний пошук: Елемент {target} знайдений за індексом {result}")
```

###  Оцінка асимптотичної складності алгоритму **лінійного пошуку** (Linear Search)

#### Завдання
Реалізувати алгоритм лінійного пошуку. Оцінити його асимптотичну складність у **O-нотації** та візуалізувати час виконання на графіку.


```python
import time
import random
import matplotlib.pyplot as plt

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

sizes = [1000, 2000, 4000, 8000, 16000, 32000, 64000]
times = []

for size in sizes:
    data = list(range(size))
    target = -1  # гарантовано не знайдеться
    
    start = time.time()
    linear_search(data, target)
    end = time.time()
    
    times.append(end - start)

plt.figure(figsize=(10, 5))
plt.plot(sizes, times, marker='o', color='blue', label="Час виконання")
plt.title("Лінійний пошук: розмір вхідних даних vs час")
plt.xlabel("Розмір масиву")
plt.ylabel("Час (секунди)")
plt.grid(True)
plt.legend()
plt.show()
```

### 📈 Асимптотична складність

- **Найгірший випадок**: елемент не знайдений ⇒ перегляд усіх `n` елементів  
- **Часова складність**: `O(n)`  
- **Просторова складність**: `O(1)` (не використовує додаткової памʼяті)

### Бінарний пошук

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid  
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1 

arr = [2, 3, 5, 7, 11, 13, 17]  
target = 7
result = binary_search(arr, target)
print(f"Бінарний пошук: Елемент {target} знайдений за індексом {result}")
```

### 🔍 Оцінка асимптотичної складності алгоритму **бінарного пошуку** (Binary Search)

#### ✅ Завдання
Реалізувати алгоритм бінарного пошуку. Оцінити його асимптотичну складність у **O-нотації** та візуалізувати час виконання на графіку.


```python
import time
import random
import matplotlib.pyplot as plt

def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
            
    return -1

sizes = [1000, 2000, 4000, 8000, 16000, 32000, 64000, 128000, 256000]
times = []

for size in sizes:
    data = list(range(size))
    target = -1  # не існує в масиві
    
    start = time.time()
    binary_search(data, target)
    end = time.time()
    
    times.append(end - start)

plt.figure(figsize=(10, 5))
plt.plot(sizes, times, marker='o', color='green', label="Час виконання")
plt.title("Бінарний пошук: розмір вхідних даних vs час")
plt.xlabel("Розмір масиву")
plt.ylabel("Час (секунди)")
plt.grid(True)
plt.legend()
plt.show()
```

### 📈 Асимптотична складність

- **Найгірший випадок**: поділ навпіл до знаходження або виключення елементу
- **Часова складність**: `O(log n)`  
- **Просторова складність**: `O(1)` (ітеративна версія без рекурсії)


# Підсумки лабораторної роботи

### 1. Визначення задачі пошуку

Задача пошуку полягає в знаходженні елемента в колекції даних, такій як масив, список, дерево або інші структури даних. Алгоритми пошуку забезпечують ефективний пошук за допомогою різних методів.

### 2. Лінійний та бінарний пошук

- **Лінійний пошук** є простим, але менш ефективним для великих наборів даних, оскільки його складність становить O(N). Це означає, що з кожним збільшенням розміру масиву час виконання лінійного пошуку збільшується пропорційно кількості елементів.
  
- **Бінарний пошук** набагато ефективніший, оскільки має складність O(log N), що дозволяє йому працювати набагато швидше на великих відсортованих даних. Він застосовується тільки до відсортованих масивів і скорочує простір пошуку вдвічі на кожному кроці.

### 3. Переваги та недоліки

- **Переваги бінарного пошуку**:
  - Швидкість, завдяки O(log N) складності.
  - Підходить для великих відсортованих даних.
  
- **Недоліки бінарного пошуку**:
  - Потрібно, щоб дані були відсортованими.
  - Складність реалізації порівняно з лінійним пошуком.
  
- **Переваги лінійного пошуку**:
  - Простота реалізації.
  - Підходить для невідсортованих даних.
  
- **Недоліки лінійного пошуку**:
  - Низька ефективність на великих даних.

### 4. Інші алгоритми пошуку

Окрім лінійного та бінарного пошуку, існують і інші варіанти, такі як:
- **Блочний пошук**.
- **Індексований пошук**.
- **Пошук у дереві**.

### 5. Висновки

- Лінійний пошук простий у реалізації, але повільний при великих масивах даних.
- Бінарний пошук значно швидший для відсортованих даних і є кращим вибором для великих наборів даних.
- Задача вибору між лінійним і бінарним пошуком залежить від типу та розміру даних. Для відсортованих даних найкраще використовувати бінарний пошук.
